```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv("Steelers.csv")
df.head()
```

```python
sns.histplot(df["Rushing Yards"], kde=False)
```

```python
sns.scatterplot(x="Rushing Yards", y="Total Yards", data=df)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
X = df[["Rushing Yards", "Total Yards"]]
y = df["Result"]
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)

knn= KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Create new data for prediction
new_data = pd.DataFrame([[125, 367]], columns=["Rushing Yards", "Total Yards"])
prediction = knn.predict(new_data)
print(prediction)

```

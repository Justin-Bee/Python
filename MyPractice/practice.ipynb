```python
def even_odd(x):
    if x % 2 == 0:
        print("Even")
    else:
        print("Odd") 

even_odd(15)           
```

```python
def max_of_two(a:int, b:int)-> int:
    if a > b:
        return a
    elif b > a:
        return b
    else:
        return 0

max_of_two(2,1234)        
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("employees.csv")

df.head()
```

```python
sns.histplot(data=df["Salary"], kde=False)
```

```python
sns.scatterplot(x="Salary", y = "Experience", data=df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor

X = df[["Salary"]]
y = df["Experience"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)

new_data = pd.DataFrame([[100000]], columns=["Salary"])  # Make sure to keep column names
prediction = knn.predict(new_data)

print(prediction[0])

print("Predicted Experience:", prediction[0])
```

```python
from sklearn.neighbors import KNeighborsClassifier
X = df[["Age", "Experience"]]
y = df["Salary"]
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
# Create new data for prediction
new_data = pd.DataFrame([[35, 4]], columns=["Age", "Experience"])

# Make prediction
prediction = knn.predict(new_data)
print(prediction[0])
print("Predicted Salary:", prediction[0])
```
